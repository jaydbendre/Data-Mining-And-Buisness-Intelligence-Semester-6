```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics.cluster import homogeneity_score

kmeans = KMeans(n_clusters = 3 )
data = pd.DataFrame(pd.read_csv('../Datasets/FinalMergedDataset/cleaned_numeric_dataset.csv'))
points = np.array(data[["Destination","Delay"]])
kmeans.fit(points)
y_km = kmeans.fit_predict(points)

plt.scatter(points[y_km ==0,0], points[y_km == 0,1], s=100, c='red')
plt.scatter(points[y_km ==1,0], points[y_km == 1,1], s=100, c='green')
plt.scatter(points[y_km ==2,0], points[y_km == 2,1], s=100, c='blue')

print(kmeans.cluster_centers_)

#Sum of squared distances of samples to their closest cluster center
print(kmeans.inertia_)

print(kmeans.score(points))
```
